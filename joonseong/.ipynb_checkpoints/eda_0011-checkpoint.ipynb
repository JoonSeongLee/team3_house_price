```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
maindf = pd.read_csv('../train.csv')
maindf.tail()
```

```python
df0011 = maindf.iloc[:, 0:11]
df0011["SalePrice"] = maindf["SalePrice"]
print(df0011.columns)
df0011.tail()
```

```python
df0011_LotConfig = pd.DataFrame(df0011, columns=['LotConfig'])
df0011_LotConfig['SalePrice'] = df0011['SalePrice'] 
df0011_f2 = df0011.groupby(['LotConfig']).size().reset_index(name='Counts')
df0011_f2   # 지하실 높이 평가
```

```python
def LotConfiggrade(data) :
    if data == 'Corner':
        return 5
    if data == 'CulDSac':
        return 4
    if data == 'FR2':
        return 3
    if data == 'FR3':
        return 2
    if data == 'Inside':
        return 1
    else:
        return 0
df0011_LotConfig['LotConfigGrade'] = df0011_LotConfig['LotConfig'].apply(LotConfiggrade)
sns.jointplot(x="LotConfigGrade", y="SalePrice", data=df0011_LotConfig)
```
