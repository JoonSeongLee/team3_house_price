```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
maindf = pd.read_csv('../train.csv')
maindf.tail()
```

```python
df5161 = maindf.iloc[:, 51:61]
df5161["SalePrice"] = maindf["SalePrice"]
print(df5161.columns)
df5161.tail()
```

```python
df5161_GarageFinish = pd.DataFrame(df5161, columns=['GarageFinish'])
df5161_GarageFinish['SalePrice'] = df5161['SalePrice'] 
df5161_f2 = df5161.groupby(['GarageFinish']).size().reset_index(name='Counts')
df5161_f2   # 차고 내부 마무리
#        Fin	Finished
#        RFn	Rough Finished
#        Unf	Unfinished
#        NA	No Garage
```

```python
def GarageFinishgrade(data) :
    if data == 'Fin':
        return 3
    if data == 'RFn':
        return 2
    if data == 'Unf':
        return 1
    else:
        return 0
df5161_GarageFinish['GarageFinishGrade'] = df5161_GarageFinish['GarageFinish'].apply(GarageFinishgrade)
sns.jointplot(x="GarageFinishGrade", y="SalePrice", data=df5161_GarageFinish)
```

```python
df5161_GarageType = pd.DataFrame(df5161, columns=['GarageType'])
df5161_GarageType['SalePrice'] = df5161['SalePrice'] 
df5161_f2 = df5161.groupby(['GarageType']).size().reset_index(name='Counts')
df5161_f2   #차고 위치
#        2Types More than one type of garage
#        Attchd Attached to home
#        Basment Basement Garage
#        BuiltIn Built-In (Garage part of house - typically has room above garage)
#        CarPort Car Port
#        Detchd Detached from home
#        NA No Garage
```

```python
def GarageTypegrade(data) :
    if data == '2Types':
        return 6
    if data == 'Attchd':
        return 5
    if data == 'Basment':
        return 4
    if data == 'BuiltIn':
        return 3
    if data == 'CarPort':
        return 2
    if data == 'Detchd':
        return 1
    else:
        return 0
df5161_GarageType['GarageTypeGrade'] = df5161_GarageType['GarageType'].apply(GarageTypegrade)
sns.jointplot(x="GarageTypeGrade", y="SalePrice", data=df5161_GarageType)
```

```python
df5161_FireplaceQu = pd.DataFrame(df5161, columns=['FireplaceQu'])
df5161_FireplaceQu['SalePrice'] = df5161['SalePrice'] 
df5161_f2 = df5161.groupby(['FireplaceQu']).size().reset_index(name='Counts')
df5161_f2   #난로 상태
#        Ex Excellent
#        Gd Good
#        TA Typical/Average
#        Fa Fair
#        Po Poor
#        NA No Garage
```

```python
def FireplaceQugrade(data) :
    if data == 'Ex':
        return 5
    if data == 'Fa':
        return 4
    if data == 'Gd':
        return 3
    if data == 'Po':
        return 2
    if data == 'TA':
        return 1
    else:
        return 0
df5161_FireplaceQu['FireplaceQuGrade'] = df5161_FireplaceQu['FireplaceQu'].apply(FireplaceQugrade)
sns.jointplot(x="FireplaceQuGrade", y="SalePrice", data=df5161_FireplaceQu)
```

```python
df5161_Functional = pd.DataFrame(df5161, columns=['Functional'])
df5161_Functional['SalePrice'] = df5161['SalePrice'] 
df5161_f2 = df5161.groupby(['Functional']).size().reset_index(name='Counts')
df5161_f2  

#집의 기능
#        Typ  Typical Functionality
#        Min1 Minor Deductions 1
#        Min2 Minor Deductions 2
#        Mod  Moderate Deductions
#        Maj1 Major Deductions 1
#        Maj2 Major Deductions 2
#        Sev  Severely Damaged
#        Sal  Salvage only
```

```python
def Functionalgrade(data) :
    if data == 'Maj1':
        return 7
    if data == 'Maj2':
        return 6    
    if data == 'Min1':
        return 5
    if data == 'Min2':
        return 4
    if data == 'Mod':
        return 3
    if data == 'Sev':
        return 2
    if data == 'Typ':
        return 1
    else:
        return 0
df5161_Functional['FunctionalGrade'] = df5161_Functional['Functional'].apply(Functionalgrade)
sns.jointplot(x="FunctionalGrade", y="SalePrice", data=df5161_Functional)
```

```python
df5161_KitchenQual = pd.DataFrame(df5161, columns=['KitchenQual'])
df5161_KitchenQual['SalePrice'] = df5161['SalePrice'] 
df5161_f2 = df5161.groupby(['KitchenQual']).size().reset_index(name='Counts')
df5161_f2  

# KitchenQual: Kitchen quality

#        Ex Excellent
#        Gd Good
#        TA Typical/Average
#        Fa Fair
#        Po Poor
```

```python
def KitchenQualgrade(data) :
    if data == 'Ex':
        return 5
    if data == 'Gd':
        return 4
    if data == 'TA':
        return 3
    if data == 'Fa':
        return 2
    if data == 'Po':
        return 1
    else:
        return 0
df5161_KitchenQual['KitchenQualGrade'] = df5161_KitchenQual['KitchenQual'].apply(KitchenQualgrade)
sns.jointplot(x="KitchenQualGrade", y="SalePrice", data=df5161_KitchenQual)
```
