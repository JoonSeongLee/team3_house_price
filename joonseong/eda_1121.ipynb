```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
maindf = pd.read_csv('../train.csv')
maindf.tail()
```

```python
df1121 = maindf.iloc[:, 11:21]
df1121["SalePrice"] = maindf["SalePrice"]
print(df1121.columns)
df1121.tail()
```

```python
df1121_HouseStyle = pd.DataFrame(df1121, columns=['HouseStyle'])
df1121_HouseStyle['SalePrice'] = df1121['SalePrice'] 
df1121_f2 = df1121.groupby(['HouseStyle']).size().reset_index(name='Counts')
df1121_f2  
```

```python
def HouseStylegrade(data) :
    if data == '1.5Fin':
        return 8
    if data == '1.5Unf':
        return 7
    if data == '1Story':
        return 6
    if data == '2.5Fin':
        return 5
    if data == '2.5Unf':
        return 4
    if data == '2Story':
        return 3
    if data == 'SFoyer':
        return 2
    if data == 'SLvl':
        return 1
    else:
        return 0
df1121_HouseStyle['HouseStyle'] = df1121_HouseStyle['HouseStyle'].apply(HouseStylegrade)
sns.jointplot(x="HouseStyle", y="SalePrice", data=df1121_HouseStyle)
```
