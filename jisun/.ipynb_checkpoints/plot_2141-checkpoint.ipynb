```python
df = df_raw
```

```python
ls = [21,22,23,24,25,27,28,29,30,31,32,33,35,39,40,80]
df1 = df.iloc[:,ls]
df1.head()
```

```python
# plot_name 0 : violinplot, 1: boxplot
def show_plot(df,plot_name):
    col_names = df.columns[0:-1]
    for col_name in col_names:
        if plot_name == 0:
            sns.violinplot(x=col_name, y='SalePrice', data=df1)
        elif plot_name == 1:
            sns.boxplot(x=col_name, y='SalePrice', data=df1)
        plt.show()
```

```python
show_plot(df1,0)
```

```python
show_plot(df1,1)
```
